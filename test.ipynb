```python
def format_flight_from_graphQLSearchResult(origin_response:requests.models.Response)-> list[flight]:
    tmp_json = origin_response.json()
    list_result = []
    match tmp_json:
        case {'data': {'appFlightSearchResult': {'id': _, 'sessionId': _, 'flightType': 'roundTrip', 'journeys': [
            {'legs': l1}, 
            {'legs': l2}
        ]}}}    if type(l1) == list and type(l2) == list and (len(l1) == 0 or 1) and (len(l2) == 0 or 1):    
            tmp_journeys = tmp_json['data']['appFlightSearchResult']['journeys']
            for i in range(2):
                tmp_legs = tmp_journeys[i]['legs']
                match tmp_legs:
                    case [{'origin': _, 'destination': _, 'departureDate': _, 'availabilityLegs': list()}]:
                        tmp_availabilityLegs = tmp_legs[0]['availabilityLegs']
                        for i in tmp_availabilityLegs:
                            f = flight()
                            f._data_insert(i)
                            list_result.append(f)
                    case []:
                        pass
                    case _ :
                        print('legs format error')
        case _ :
            print('graphQLSearchResult.josn format error')
    return list_result
```

```python
lst_flights:list[flight] = []
for i in list_m:
    for j in i:
        lst = format_flight_from_graphQLSearchResult(j)
        lst_flights += lst
print(len(lst_flights))
```

```python
dict_flights:dict[str,list[flight]] = {}
for i in lst_flights:
    tmp_key = i.origin + '_' + i.destination
    if tmp_key in dict_flights:
        dict_flights[tmp_key].append(i)
    else :
        dict_flights[tmp_key] = [i]
```
