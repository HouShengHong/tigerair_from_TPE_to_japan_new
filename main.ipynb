```python
def initialize_driver(driver:tigerair_webdriver,wait:int):
    time.sleep(wait)
    driver.standard_init()

class class_driver(Enum):
    kanto:tigerair_webdriver              = tigerair_webdriver(Destination_airport.kanto)
    kansai_and_shikoku:tigerair_webdriver = tigerair_webdriver(Destination_airport.kansai_and_shikoku)
    kyushu:tigerair_webdriver             = tigerair_webdriver(Destination_airport.kyushu)
    chubu:tigerair_webdriver              = tigerair_webdriver(Destination_airport.chubu)
    tohoku:tigerair_webdriver             = tigerair_webdriver(Destination_airport.tohoku)
    hokkaido:tigerair_webdriver           = tigerair_webdriver(Destination_airport.hokkaido)

int_tmp = 0
list_process:list[Process] = []
for i in class_driver:
    p = Process(
        target=initialize_driver,
        args=(i.value,int_tmp*5)
    )
    int_tmp += 1
    list_process.append(p)
    p.start()

for i in list_process:
    i.join()
```

```python
for i in class_driver:
    print(i.value._graphQL_id_update())
    i.value.graphQL_currency_change(Currency.TWD)
    i.value.graphQL_date_change(datetime.date.today())
```

```python
def search(lst:list,driver:tigerair_webdriver,start_day:datetime.date,delta:int):
    for i in range(delta):
        print(i)
        driver.graphQL_date_change(start_day + datetime.timedelta(days=i))
        lst.append(driver.graphQL_search_result())

m_kanto = Manager().list()
m_kansai_and_shikoku = Manager().list()
m_kyushu = Manager().list()
m_chubu = Manager().list()
m_tohoku = Manager().list()
m_hokkaido = Manager().list()

list_m = [
    m_kanto,
    m_kansai_and_shikoku,
    m_kyushu,
    m_chubu,
    m_tohoku,
    m_hokkaido,
]

int_tmp = 0
list_process:list[Process] = []
for i in class_driver:
    p = Process(
        target=search,
        args=(list_m[int_tmp],i.value,datetime.date.today(),3)
    )
    int_tmp += 1
    list_process.append(p)
    p.start()

for i in list_process:
    i.join()
```
